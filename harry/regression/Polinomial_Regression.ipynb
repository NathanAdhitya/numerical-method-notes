```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6)
y = np.array([2.1,7.7,13.6,27.2,40.9,61.1])
N = np.array([
    [len(x), x.sum(), (x**2).sum()],
    [x.sum(), (x**2).sum(), (x**3).sum()],
    [(x**2).sum(), (x**3).sum(), (x**4).sum()]
])
r = np.array([y.sum(), (x*y).sum(), (x**2 * y).sum()])

a = np.linalg.solve(N, r)

x_draw = np.arange(0, 6, 0.5)
y_draw = a[0] + x_draw*a[1] + x_draw**2 * a[2]
plt.plot(x_draw, y_draw, color="#FF0000")
plt.plot(x, y, 'o', color="#00FF00")
plt.grid(True)
plt.show()
```
